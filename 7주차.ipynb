```python
import pandas as pd
import numpy as np
```

```python
exam=pd.read_csv('exam.csv')
```

```python
mpg=pd.read_csv('mpg.csv')
```

```python
copy_mpg=mpg.copy()
```

```python
copy_mpg.assign(cty_hwy_sum = copy_mpg['cty']+copy_mpg['hwy'], 
                cty_hwy_sum_mean = lambda x : x['cty_hwy_sum'].mean())
```

```python
copy_mpg.assign(cty_hwy_sum = copy_mpg['cty']+copy_mpg['hwy'], 
                cty_hwy_sum_mean = lambda x : x['cty_hwy_sum'].mean()).sort_values('cty_hwy_sum', ascending = False).head(3)
```

```python
mpg.assign(cty_hwy_sum = mpg['cty']+mpg['hwy'], 
           cty_hwy_sum_mean = lambda x : x['cty_hwy_sum'].mean())\
   .sort_values('cty_hwy_sum', ascending = False)\
   .head(3)
```
